```python
def tui_la_func_a(): return print('Hello, tui la func_a')
def tui_la_func_b(): return print('Hello, tui la func_b')
def tui_la_func_c(x): return x


def tui_la_func_yield(value):
    tui_la_func_a()
    yield tui_la_func_c(value)
    tui_la_func_b()


tui_la_func = tui_la_func_yield('Tui la giá trị truyền vào')
print("Tui la >> ", tui_la_func)

for _func in tui_la_func:
    print(_func)
```

```python
def outer_func(x):
    def inner_func(y):
        return x + y

    return inner_func


print(f"outer_func: {outer_func(10)(5)}")  # Output: 15

closure = outer_func(10)
print(f"closure   : {closure(5)}")  # Output: 15

print(">>>>>>>>>>>>>>>")
print("Secret Password")


class Secure:
    def __init__(self, data):
        self.data = data
        self.password = "secret"

    def get_data(self, passwd):
        return self.data if passwd == self.password else None


def secure(data):
    password = "secret"

    def _get_data(passwd):
        if passwd == password:
            return data

    return _get_data


print()
print(">>>> From Class")
secure_data = Secure("Aa@12345")
print(f'Passwd secret: {secure_data.get_data("secret")}')  # Output: 'Aa@12345'
# Output: None
print(f'Passwd wrongs: {secure_data.get_data("Wrong password")}')

print()
print(">>>> From closure function")
# Output: 'Aa@12345'
print(f'Passwd secret 1.1: {secure("Aa@12345")("secret")}')
# Output: None
print(f'Passwd wrongs 1.2: {secure("Aa@12345")("Wrong password")}')

secure_data_func = secure("Aa@12345")
print(f'Passwd secret 2.1: {secure_data_func("secret")}')  # Output: 'Aa@12345'
# Output: None
print(f'Passwd wrongs 2.2: {secure_data_func("Wrong password")}')
```

```python
import functools


def decorator_func(f):
    def wrapper_func():
        print()
        print("Before func is called")
        f()
        print("After func is called")

    return wrapper_func


def decorator_wraps_func(f):
    @functools.wraps(f)
    def wrapper(*args, **kwargs):
        print()
        print("Before func is called")
        result = f(*args, **kwargs)
        print("After func is called")
        return result

    return wrapper


@decorator_func
def hello_world():
    print("Hello World, >> use @decorator_func")


def __hello__():
    print("Hello World, >> use f(f)")


hello = decorator_func(__hello__)


@decorator_wraps_func
def hello_world_wrap():
    print("Hello World, >> use @decorator_wraps_func")


def __hellowrap__():
    print("Hello World, >> use wrap f(f)")


hello_wrap = decorator_wraps_func(__hellowrap__)

###
hello_world()
hello()
hello_world_wrap()
hello_wrap()
```

```python
# https://realpython.com/python-metaclasses/
# https://www.datacamp.com/tutorial/python-metaclasses
# https://viblo.asia/p/metaclass-in-python-gDVK24JnlLj

# Simple inheritance


def secret_password(data):
    password = "simple-secret"

    def secret(passwd):
        if passwd == password:
            return data

    return secret


class BaseClassSimple:
    HELLO_BASE = "Hello BaseClass"

    def __init__(self, x=None, y=None, z=None):
        self.x = x
        self.y = y
        self.z = z
        self.secret_password = secret_password("Base-Aa@12345")
        self.baseclass = "BaseClass"


class XBaseClass(BaseClassSimple):
    pass


class YBaseClass(BaseClassSimple):
    pass


class ZBaseClass(BaseClassSimple):
    pass


class VBaseClass(BaseClassSimple):
    pass


print("Class --- Simple")
print(BaseClassSimple(type("HelloBaseClass", (), {})).HELLO_BASE)
print(XBaseClass(x=100).x)
print(YBaseClass(y=200).y)
print(ZBaseClass(z=300).z)
print(VBaseClass().baseclass)
print(VBaseClass().secret_password("simple-secret"))
print(VBaseClass().secret_password("wrongs-password"))

# Class Decorator


def BaseClassDecorator(cls):
    def _secret_password(data):
        password = "decorator-secret"

        def secret(passwd):
            if passwd == password:
                return data

        return secret

    class InnerClass(cls):
        HELLO_INNER = "Hello InnerClass"

        def __init__(self, x=None, y=None, z=None):
            self.x = x
            self.y = y
            self.z = z
            self.secret_password = _secret_password("Decorator-Aa@12345")
            self.innerclass = "InnerClass"

    return InnerClass


@BaseClassDecorator
class XClass:
    pass


@BaseClassDecorator
class YClass:
    pass


@BaseClassDecorator
class ZClass:
    pass


@BaseClassDecorator
class VClass:
    pass


print()
print("Class --- Decorator")
print(BaseClassDecorator(type("HelloInnerClass", (), {})).HELLO_INNER)
print(XClass(x=100).x)
print(YClass(y=200).y)
print(ZClass(z=300).z)
print(VClass().innerclass)
print(VClass().secret_password("decorator-secret"))  # -> Decorator-Aa@12345
print(VClass().secret_password("Wrong Password"))  # -> None
```
